# Outlier Detection and Removal using Z-Score

Detect and remove outliers using Z-Score filtering. Outliers are defined as values beyond  2 from the mean. This helps improve model performance by removing extreme values that could negatively impact training.

```python
import numpy as np
from scipy import stats

# Get only numeric columns for outlier detection
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
# Remove the target column from outlier detection if it exists
if 'Result' in numeric_columns:
    numeric_columns.remove('Result')

print("Numeric columns for outlier detection:", numeric_columns)
print(f"Original dataset shape: {df.shape}")

# Store original data for comparison
df_original = df.copy()

# Calculate Z-scores for each numeric column
outlier_indices = set()

print("\nOutlier Analysis per Feature:")
print("-" * 50)

for column in numeric_columns:
    # Calculate Z-scores
    z_scores = np.abs(stats.zscore(df[column]))
    
    # Find outliers (Z-score > 2)
    column_outliers = np.where(z_scores > 2)[0]
    outlier_indices.update(column_outliers)
    
    print(f"{column}:")
    print(f"  - Mean: {df[column].mean():.2f}")
    print(f"  - Std: {df[column].std():.2f}")
    print(f"  - Outliers found: {len(column_outliers)}")
    print(f"  - Outlier percentage: {len(column_outliers)/len(df)*100:.2f}%")
    
    if len(column_outliers) > 0:
        print(f"  - Outlier values range: {df.iloc[column_outliers][column].min():.2f} to {df.iloc[column_outliers][column].max():.2f}")
    print()

print(f"Total unique outlier rows identified: {len(outlier_indices)}")
print(f"Percentage of data to be removed: {len(outlier_indices)/len(df)*100:.2f}%")
```

```python
# Visualize outliers before removal
import matplotlib.pyplot as plt

# Create box plots to visualize outliers
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for i, column in enumerate(numeric_columns[:6]):  # Show first 6 numeric columns
    if i < len(axes):
        axes[i].boxplot(df[column])
        axes[i].set_title(f'{column}\n(Outliers beyond 2)')
        axes[i].set_ylabel('Values')
        
        # Add mean and std lines
        mean_val = df[column].mean()
        std_val = df[column].std()
        axes[i].axhline(y=mean_val, color='red', linestyle='--', alpha=0.7, label='Mean')
        axes[i].axhline(y=mean_val + 2*std_val, color='orange', linestyle='--', alpha=0.7, label='+2')
        axes[i].axhline(y=mean_val - 2*std_val, color='orange', linestyle='--', alpha=0.7, label='-2')
        axes[i].legend()

# Hide empty subplots
for i in range(len(numeric_columns), len(axes)):
    axes[i].set_visible(False)

plt.tight_layout()
plt.show()

# Show distribution of outliers across features
if len(numeric_columns) > 6:
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.ravel()
    
    for i, column in enumerate(numeric_columns[6:12]):  # Show next 6 columns if they exist
        if i < len(axes):
            axes[i].boxplot(df[column])
            axes[i].set_title(f'{column}\n(Outliers beyond 2)')
            axes[i].set_ylabel('Values')
            
            # Add mean and std lines
            mean_val = df[column].mean()
            std_val = df[column].std()
            axes[i].axhline(y=mean_val, color='red', linestyle='--', alpha=0.7, label='Mean')
            axes[i].axhline(y=mean_val + 2*std_val, color='orange', linestyle='--', alpha=0.7, label='+2')
            axes[i].axhline(y=mean_val - 2*std_val, color='orange', linestyle='--', alpha=0.7, label='-2')
            axes[i].legend()
    
    # Hide empty subplots
    for i in range(len(numeric_columns[6:12]), len(axes)):
        axes[i].set_visible(False)
    
    plt.tight_layout()
    plt.show()
```

```python
# Remove outliers from the dataset
print("Removing outliers...")

# Create a clean dataset by removing outlier rows
df_clean = df.drop(index=list(outlier_indices)).reset_index(drop=True)

print(f"Original dataset shape: {df_original.shape}")
print(f"Dataset shape after outlier removal: {df_clean.shape}")
print(f"Rows removed: {len(df) - len(df_clean)}")
print(f"Percentage of data retained: {len(df_clean)/len(df)*100:.2f}%")

# Update the main dataframe
df = df_clean.copy()

print("\nOutlier removal completed successfully!")
print(f"Final dataset shape: {df.shape}")
```

# Feature Selection - Five Methods Applied

Apply comprehensive feature selection using five different techniques to identify the most important features for model training. This helps improve model performance and interpretability by removing irrelevant or redundant features.

```python
# Import necessary libraries for feature selection
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE, SelectKBest, f_classif, chi2
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Get feature names after preprocessing
if hasattr(ct, 'get_feature_names_out'):
    feature_names = ct.get_feature_names_out()
else:
    # For older sklearn versions
    feature_names = [f'feature_{i}' for i in range(X_train.shape[1])]

print(f"Total features available: {len(feature_names)}")
print(f"Feature names: {list(feature_names)}")

# Store results from all feature selection methods
feature_selection_results = {}
```

## Method 1: Pearson Correlation Analysis

Identify features with high correlation to the target variable. Features with higher absolute correlation values are considered more important.

```python
# Method 1: Pearson Correlation
print("="*60)
print("METHOD 1: PEARSON CORRELATION ANALYSIS")
print("="*60)

# Create DataFrame with features and target for correlation analysis
X_train_df = pd.DataFrame(X_train, columns=feature_names)
correlation_data = X_train_df.copy()
correlation_data['target'] = y_train

# Calculate correlation with target
correlations = correlation_data.corr()['target'].drop('target')
abs_correlations = correlations.abs().sort_values(ascending=False)

print(f"Feature correlations with target (sorted by absolute value):")
for feature, corr in abs_correlations.head(10).items():
    print(f"  {feature}: {corr:.4f}")

# Select top features based on correlation threshold
correlation_threshold = 0.1  # Adjust as needed
selected_features_corr = abs_correlations[abs_correlations >= correlation_threshold].index.tolist()

print(f"\nSelected features (correlation >= {correlation_threshold}): {len(selected_features_corr)}")
print(f"Features: {selected_features_corr}")

# Store results
feature_selection_results['Pearson_Correlation'] = {
    'selected_features': selected_features_corr,
    'scores': abs_correlations[selected_features_corr].to_dict(),
    'method_info': f'Threshold: {correlation_threshold}'
}

# Visualize correlations
plt.figure(figsize=(12, 8))
top_corr = abs_correlations.head(15)
plt.barh(range(len(top_corr)), top_corr.values)
plt.yticks(range(len(top_corr)), top_corr.index)
plt.xlabel('Absolute Correlation with Target')
plt.title('Top 15 Features by Pearson Correlation')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Method 2: Recursive Feature Elimination (RFE)

Use RFE with Logistic Regression to recursively eliminate features and select the most important ones based on model performance.

```python
# Method 2: Recursive Feature Elimination (RFE)
print("="*60)
print("METHOD 2: RECURSIVE FEATURE ELIMINATION (RFE)")
print("="*60)

# Use Logistic Regression as the estimator for RFE
lr_estimator = LogisticRegression(random_state=42, max_iter=1000)

# Select top k features using RFE
k_features = min(10, X_train.shape[1])  # Select top 10 or all features if less than 10
rfe = RFE(estimator=lr_estimator, n_features_to_select=k_features)
rfe.fit(X_train, y_train)

# Get selected features
selected_features_rfe = [feature_names[i] for i in range(len(feature_names)) if rfe.support_[i]]
feature_rankings = {feature_names[i]: rfe.ranking_[i] for i in range(len(feature_names))}

print(f"Selected features by RFE (top {k_features}): {len(selected_features_rfe)}")
print(f"Features: {selected_features_rfe}")

print(f"\nFeature Rankings (1 = most important):")
sorted_rankings = sorted(feature_rankings.items(), key=lambda x: x[1])
for feature, rank in sorted_rankings[:15]:
    status = " Selected" if feature in selected_features_rfe else " Eliminated"
    print(f"  {feature}: Rank {rank} - {status}")

# Store results
feature_selection_results['RFE'] = {
    'selected_features': selected_features_rfe,
    'scores': {f: 1/rank for f, rank in feature_rankings.items()},  # Convert rank to score
    'method_info': f'Top {k_features} features selected'
}

# Visualize RFE rankings
plt.figure(figsize=(12, 8))
sorted_features = sorted(feature_rankings.keys(), key=lambda x: feature_rankings[x])[:15]
ranks = [feature_rankings[f] for f in sorted_features]
colors = ['green' if f in selected_features_rfe else 'gray' for f in sorted_features]

plt.barh(range(len(sorted_features)), [1/r for r in ranks], color=colors)
plt.yticks(range(len(sorted_features)), sorted_features)
plt.xlabel('Feature Importance (1/Rank)')
plt.title('Top 15 Features by RFE Ranking')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Method 3: SelectKBest with ANOVA F-value

Use ANOVA F-test to select features based on their statistical significance in distinguishing between classes.

```python
# Method 3: SelectKBest with ANOVA F-value
print("="*60)
print("METHOD 3: SELECTKBEST WITH ANOVA F-VALUE")
print("="*60)

# Apply SelectKBest with f_classif (ANOVA F-test)
k_best = min(10, X_train.shape[1])
selector_fvalue = SelectKBest(score_func=f_classif, k=k_best)
X_train_fvalue = selector_fvalue.fit_transform(X_train, y_train)

# Get selected features and their F-values
selected_features_fvalue = [feature_names[i] for i in range(len(feature_names)) if selector_fvalue.get_support()[i]]
f_scores = {feature_names[i]: selector_fvalue.scores_[i] for i in range(len(feature_names))}

print(f"Selected features by ANOVA F-value (top {k_best}): {len(selected_features_fvalue)}")
print(f"Features: {selected_features_fvalue}")

print(f"\nTop F-values:")
sorted_f_scores = sorted(f_scores.items(), key=lambda x: x[1], reverse=True)
for feature, score in sorted_f_scores[:15]:
    status = " Selected" if feature in selected_features_fvalue else " Not selected"
    print(f"  {feature}: F-value = {score:.4f} - {status}")

# Store results
feature_selection_results['ANOVA_F_value'] = {
    'selected_features': selected_features_fvalue,
    'scores': {f: score for f, score in f_scores.items() if f in selected_features_fvalue},
    'method_info': f'Top {k_best} features by F-value'
}

# Visualize F-values
plt.figure(figsize=(12, 8))
top_f_scores = dict(sorted_f_scores[:15])
colors = ['green' if f in selected_features_fvalue else 'gray' for f in top_f_scores.keys()]

plt.barh(range(len(top_f_scores)), list(top_f_scores.values()), color=colors)
plt.yticks(range(len(top_f_scores)), list(top_f_scores.keys()))
plt.xlabel('ANOVA F-value')
plt.title('Top 15 Features by ANOVA F-value')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Method 4: Chi-Square Test

Apply Chi-Square test for feature selection. Note: Chi-Square requires non-negative features, so we'll use absolute values or apply appropriate transformations.

```python
# Method 4: Chi-Square Test
print("="*60)
print("METHOD 4: CHI-SQUARE TEST")
print("="*60)

# Chi-square requires non-negative values, so we'll use MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Transform features to be non-negative for Chi-square test
min_max_scaler = MinMaxScaler()
X_train_chi = min_max_scaler.fit_transform(X_train)

# Apply SelectKBest with chi2
k_chi = min(10, X_train.shape[1])
selector_chi2 = SelectKBest(score_func=chi2, k=k_chi)

try:
    X_train_chi2 = selector_chi2.fit_transform(X_train_chi, y_train)
    
    # Get selected features and their Chi-square values
    selected_features_chi2 = [feature_names[i] for i in range(len(feature_names)) if selector_chi2.get_support()[i]]
    chi2_scores = {feature_names[i]: selector_chi2.scores_[i] for i in range(len(feature_names))}
    
    print(f"Selected features by Chi-Square (top {k_chi}): {len(selected_features_chi2)}")
    print(f"Features: {selected_features_chi2}")
    
    print(f"\nTop Chi-Square values:")
    sorted_chi2_scores = sorted(chi2_scores.items(), key=lambda x: x[1], reverse=True)
    for feature, score in sorted_chi2_scores[:15]:
        status = " Selected" if feature in selected_features_chi2 else " Not selected"
        print(f"  {feature}: Chi = {score:.4f} - {status}")
    
    # Store results
    feature_selection_results['Chi_Square'] = {
        'selected_features': selected_features_chi2,
        'scores': {f: score for f, score in chi2_scores.items() if f in selected_features_chi2},
        'method_info': f'Top {k_chi} features by Chi-Square'
    }
    
    # Visualize Chi-square values
    plt.figure(figsize=(12, 8))
    top_chi2_scores = dict(sorted_chi2_scores[:15])
    colors = ['green' if f in selected_features_chi2 else 'gray' for f in top_chi2_scores.keys()]
    
    plt.barh(range(len(top_chi2_scores)), list(top_chi2_scores.values()), color=colors)
    plt.yticks(range(len(top_chi2_scores)), list(top_chi2_scores.keys()))
    plt.xlabel('Chi-Square Score')
    plt.title('Top 15 Features by Chi-Square Test')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"Chi-Square test failed: {e}")
    print("This might be due to data characteristics. Skipping Chi-Square method.")
    feature_selection_results['Chi_Square'] = {
        'selected_features': [],
        'scores': {},
        'method_info': 'Method failed - likely due to data characteristics'
    }
```

## Method 5: Extra Trees Classifier Feature Importance

Use Extra Trees Classifier to calculate feature importance based on the reduction in impurity when splitting on each feature.

```python
# Method 5: Extra Trees Classifier Feature Importance
print("="*60)
print("METHOD 5: EXTRA TREES CLASSIFIER FEATURE IMPORTANCE")
print("="*60)

# Train Extra Trees Classifier for feature importance
et_selector = ExtraTreesClassifier(n_estimators=100, random_state=42, n_jobs=-1)
et_selector.fit(X_train, y_train)

# Get feature importances
feature_importances = {feature_names[i]: et_selector.feature_importances_[i] for i in range(len(feature_names))}

# Select top features based on importance threshold
importance_threshold = 0.05  # Adjust as needed
selected_features_et = [f for f, imp in feature_importances.items() if imp >= importance_threshold]

# If threshold results in too few features, select top k features
if len(selected_features_et) < 5:
    k_et = min(10, X_train.shape[1])
    sorted_importances = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True)
    selected_features_et = [f for f, _ in sorted_importances[:k_et]]

print(f"Selected features by Extra Trees (importance >= {importance_threshold}): {len(selected_features_et)}")
print(f"Features: {selected_features_et}")

print(f"\nTop Feature Importances:")
sorted_importances = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True)
for feature, importance in sorted_importances[:15]:
    status = " Selected" if feature in selected_features_et else " Not selected"
    print(f"  {feature}: Importance = {importance:.4f} - {status}")

# Store results
feature_selection_results['Extra_Trees'] = {
    'selected_features': selected_features_et,
    'scores': {f: imp for f, imp in feature_importances.items() if f in selected_features_et},
    'method_info': f'Importance threshold: {importance_threshold}'
}

# Visualize feature importances
plt.figure(figsize=(12, 8))
top_importances = dict(sorted_importances[:15])
colors = ['green' if f in selected_features_et else 'gray' for f in top_importances.keys()]

plt.barh(range(len(top_importances)), list(top_importances.values()), color=colors)
plt.yticks(range(len(top_importances)), list(top_importances.keys()))
plt.xlabel('Feature Importance')
plt.title('Top 15 Features by Extra Trees Classifier Importance')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Feature Selection Summary and Consensus

Compare results from all five methods and identify consensus features that are selected by multiple methods.

```python
# Feature Selection Summary and Consensus Analysis
print("="*80)
print("FEATURE SELECTION SUMMARY - CONSENSUS ANALYSIS")
print("="*80)

# Create a comprehensive summary
all_features = set()
for method, results in feature_selection_results.items():
    all_features.update(results['selected_features'])

# Count how many methods selected each feature
feature_votes = {}
for feature in all_features:
    votes = 0
    methods_selected = []
    for method, results in feature_selection_results.items():
        if feature in results['selected_features']:
            votes += 1
            methods_selected.append(method)
    feature_votes[feature] = {'votes': votes, 'methods': methods_selected}

# Sort features by consensus (number of votes)
consensus_features = sorted(feature_votes.items(), key=lambda x: x[1]['votes'], reverse=True)

print("Feature Selection Results by Method:")
print("-" * 50)
for method, results in feature_selection_results.items():
    print(f"{method}: {len(results['selected_features'])} features selected")
    print(f"  {results['method_info']}")
    print()

print("Consensus Ranking (Features selected by multiple methods):")
print("-" * 60)
for feature, info in consensus_features:
    if info['votes'] > 1:
        print(f"{feature}:")
        print(f"  Selected by {info['votes']}/{len(feature_selection_results)} methods")
        print(f"  Methods: {', '.join(info['methods'])}")
        print()

# Select final features based on consensus
# Features selected by at least 2 methods
final_consensus_features = [f for f, info in consensus_features if info['votes'] >= 2]

# If too few consensus features, add top features from individual methods
if len(final_consensus_features) < 5:
    # Add top features from Extra Trees (usually most reliable)
    et_features = feature_selection_results['Extra_Trees']['selected_features']
    for feature in et_features:
        if feature not in final_consensus_features and len(final_consensus_features) < 10:
            final_consensus_features.append(feature)

print(f"FINAL CONSENSUS FEATURES ({len(final_consensus_features)} selected):")
print("-" * 60)
for i, feature in enumerate(final_consensus_features, 1):
    votes = feature_votes[feature]['votes']
    methods = ', '.join(feature_votes[feature]['methods'])
    print(f"{i:2d}. {feature} (voted by {votes} methods: {methods})")

# Store final results
feature_selection_results['Final_Consensus'] = {
    'selected_features': final_consensus_features,
    'method_info': f'{len(final_consensus_features)} features selected by consensus'
}
```

```python
# Visualize Feature Selection Results
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Consensus Heatmap
consensus_matrix = []
methods_list = list(feature_selection_results.keys())[:-1]  # Exclude 'Final_Consensus'
top_features = [f for f, _ in consensus_features[:20]]  # Top 20 features

for feature in top_features:
    row = []
    for method in methods_list:
        if feature in feature_selection_results[method]['selected_features']:
            row.append(1)
        else:
            row.append(0)
    consensus_matrix.append(row)

sns.heatmap(consensus_matrix, 
            xticklabels=methods_list,
            yticklabels=top_features,
            annot=True, 
            cmap='RdYlGn', 
            ax=axes[0,0])
axes[0,0].set_title('Feature Selection Consensus Heatmap\n(1=Selected, 0=Not Selected)')
axes[0,0].set_xlabel('Selection Methods')
axes[0,0].set_ylabel('Features')

# 2. Number of features selected by each method
method_counts = [len(results['selected_features']) for results in feature_selection_results.values()][:-1]
axes[0,1].bar(methods_list, method_counts, color=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightgray'])
axes[0,1].set_title('Number of Features Selected by Each Method')
axes[0,1].set_ylabel('Number of Features')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Consensus vote distribution
vote_counts = {}
for feature, info in feature_votes.items():
    votes = info['votes']
    vote_counts[votes] = vote_counts.get(votes, 0) + 1

votes_list = sorted(vote_counts.keys())
counts_list = [vote_counts[v] for v in votes_list]
axes[1,0].bar(votes_list, counts_list, color='lightblue')
axes[1,0].set_title('Distribution of Consensus Votes')
axes[1,0].set_xlabel('Number of Methods Selecting Feature')
axes[1,0].set_ylabel('Number of Features')

# 4. Final consensus features
if len(final_consensus_features) <= 15:
    axes[1,1].barh(range(len(final_consensus_features)), 
                   [feature_votes[f]['votes'] for f in final_consensus_features],
                   color='green')
    axes[1,1].set_yticks(range(len(final_consensus_features)))
    axes[1,1].set_yticklabels(final_consensus_features)
    axes[1,1].set_xlabel('Number of Methods')
    axes[1,1].set_title('Final Consensus Features')
    axes[1,1].invert_yaxis()
else:
    # Show top 15 if too many features
    top_consensus = final_consensus_features[:15]
    axes[1,1].barh(range(len(top_consensus)), 
                   [feature_votes[f]['votes'] for f in top_consensus],
                   color='green')
    axes[1,1].set_yticks(range(len(top_consensus)))
    axes[1,1].set_yticklabels(top_consensus)
    axes[1,1].set_xlabel('Number of Methods')
    axes[1,1].set_title('Top 15 Final Consensus Features')
    axes[1,1].invert_yaxis()

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("FEATURE SELECTION COMPLETE")
print("="*80)
print(f" Applied 5 feature selection methods")
print(f" Identified {len(final_consensus_features)} consensus features")
print(f" Ready for model training with selected features")
print(f" Feature selection results stored in 'feature_selection_results' dictionary")
```

## 1. Stacking Ensemble Model

## 2. Convolutional Neural Network (CNN)

Apply CNN to tabular data by reshaping features into a 2D grid format. CNNs can capture local patterns in the feature space.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import math
import numpy as np

print("="*60)
print("CONVOLUTIONAL NEURAL NETWORK (CNN)")
print("="*60)

class CNN1D(nn.Module):
    def __init__(self, input_features):
        super(CNN1D, self).__init__()
        
        # For tabular data, we'll use 1D convolutions
        self.conv_layers = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Dropout(0.2),
            
            nn.Conv1d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.Dropout(0.2),
            
            nn.Conv1d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm1d(128),
            nn.Dropout(0.2)
        )
        
        # Store input features
        self.input_features = input_features
        
        # Classifier with global pooling output size
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 256),  # After global pooling, we have 128 features
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        # Reshape for 1D convolution: (batch_size, 1, features)
        x = x.unsqueeze(1)
        x = self.conv_layers(x)
        # Manual global average pooling (MPS compatible)
        x = torch.mean(x, dim=2, keepdim=False)  # Global average pooling, remove the sequence dimension
        x = self.classifier(x)
        return x

# Prepare data
device = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

X_train_cnn = torch.FloatTensor(X_train).to(device)
y_train_cnn = torch.FloatTensor(y_train).unsqueeze(1).to(device)
X_test_cnn = torch.FloatTensor(X_test).to(device)
y_test_cnn = torch.FloatTensor(y_test).unsqueeze(1).to(device)

# Create model
cnn_model = CNN1D(X_train.shape[1]).to(device)

# Training setup
criterion = nn.BCELoss()
optimizer = optim.Adam(cnn_model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=10, factor=0.5)

# Data loaders
train_dataset_cnn = TensorDataset(X_train_cnn, y_train_cnn)
train_loader_cnn = DataLoader(train_dataset_cnn, batch_size=32, shuffle=True)

# Training loop
num_epochs = 100
train_losses_cnn = []
print("Training CNN...")

for epoch in range(num_epochs):
    cnn_model.train()
    epoch_loss = 0.0
    
    for batch_X, batch_y in train_loader_cnn:
        optimizer.zero_grad()
        outputs = cnn_model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader_cnn)
    train_losses_cnn.append(avg_loss)
    scheduler.step(avg_loss)
    
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')

# Evaluation
cnn_model.eval()
with torch.no_grad():
    test_outputs_cnn = cnn_model(X_test_cnn)
    test_proba_cnn = test_outputs_cnn.cpu().numpy().ravel()
    test_pred_cnn = (test_proba_cnn >= 0.5).astype(int)

# Calculate metrics
y_test_np = y_test_cnn.cpu().numpy().ravel()
accuracy_cnn = accuracy_score(y_test_np, test_pred_cnn)
auc_cnn = roc_auc_score(y_test_np, test_proba_cnn)

print(f"\nCNN Results:")
print(f"Accuracy: {accuracy_cnn:.4f}")
print(f"ROC-AUC: {auc_cnn:.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test_np, test_pred_cnn))
print("\nClassification Report:")
print(classification_report(y_test_np, test_pred_cnn))

# Plot results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Training loss
ax1.plot(train_losses_cnn)
ax1.set_title('CNN Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')

# ROC curve
fpr_cnn, tpr_cnn, _ = roc_curve(y_test_np, test_proba_cnn)
ax2.plot(fpr_cnn, tpr_cnn, label=f'CNN (AUC = {auc_cnn:.3f})')
ax2.plot([0, 1], [0, 1], 'k--')
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('ROC Curve - CNN')
ax2.legend()

plt.tight_layout()
plt.show()
```

## 5. Bidirectional LSTM (Bi-LSTM)

Apply Bi-LSTM to tabular data by treating each feature as a time step, allowing the model to capture sequential patterns in both directions.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

print("="*60)
print("BIDIRECTIONAL LSTM (Bi-LSTM)")
print("="*60)

class BiLSTM(nn.Module):
    def __init__(self, input_dim):
        super(BiLSTM, self).__init__()
        
        # For tabular data, we'll treat each feature as a sequence element
        self.lstm = nn.LSTM(
            input_size=1,  # Each feature is treated as a single value
            hidden_size=64,
            num_layers=2,
            batch_first=True,
            dropout=0.2,
            bidirectional=True
        )
        
        # After bidirectional LSTM: hidden_size * 2 * input_dim
        lstm_output_size = 64 * 2  # bidirectional doubles the output size
        
        self.classifier = nn.Sequential(
            nn.Linear(lstm_output_size, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        # Reshape: (batch_size, features) -> (batch_size, features, 1)
        x = x.unsqueeze(-1)
        
        # LSTM forward pass
        lstm_out, (hidden, cell) = self.lstm(x)
        
        # Take the last output from both directions
        # lstm_out shape: (batch_size, seq_len, hidden_size * 2)
        lstm_out = lstm_out[:, -1, :]  # Take last time step
        
        # Classification
        output = self.classifier(lstm_out)
        return output

# Prepare data
device = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

X_train_bilstm = torch.FloatTensor(X_train).to(device)
y_train_bilstm = torch.FloatTensor(y_train).unsqueeze(1).to(device)
X_test_bilstm = torch.FloatTensor(X_test).to(device)
y_test_bilstm = torch.FloatTensor(y_test).unsqueeze(1).to(device)

# Create model
bilstm_model = BiLSTM(X_train.shape[1]).to(device)

# Training setup
criterion = nn.BCELoss()
optimizer = optim.Adam(bilstm_model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=10, factor=0.5)

# Data loader
train_dataset_bilstm = TensorDataset(X_train_bilstm, y_train_bilstm)
train_loader_bilstm = DataLoader(train_dataset_bilstm, batch_size=32, shuffle=True)

# Training loop
num_epochs = 100
train_losses_bilstm = []
print("Training Bi-LSTM...")

for epoch in range(num_epochs):
    bilstm_model.train()
    epoch_loss = 0.0
    
    for batch_X, batch_y in train_loader_bilstm:
        optimizer.zero_grad()
        outputs = bilstm_model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        
        # Gradient clipping for LSTM
        torch.nn.utils.clip_grad_norm_(bilstm_model.parameters(), max_norm=1.0)
        
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader_bilstm)
    train_losses_bilstm.append(avg_loss)
    scheduler.step(avg_loss)
    
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')

# Evaluation
bilstm_model.eval()
with torch.no_grad():
    test_outputs_bilstm = bilstm_model(X_test_bilstm)
    test_proba_bilstm = test_outputs_bilstm.cpu().numpy().ravel()
    test_pred_bilstm = (test_proba_bilstm >= 0.5).astype(int)

# Calculate metrics
y_test_np = y_test_bilstm.cpu().numpy().ravel()
accuracy_bilstm = accuracy_score(y_test_np, test_pred_bilstm)
auc_bilstm = roc_auc_score(y_test_np, test_proba_bilstm)

print(f"\nBi-LSTM Results:")
print(f"Accuracy: {accuracy_bilstm:.4f}")
print(f"ROC-AUC: {auc_bilstm:.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test_np, test_pred_bilstm))
print("\nClassification Report:")
print(classification_report(y_test_np, test_pred_bilstm))

# Plot results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Training loss
ax1.plot(train_losses_bilstm)
ax1.set_title('Bi-LSTM Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')

# ROC curve
fpr_bilstm, tpr_bilstm, _ = roc_curve(y_test_np, test_proba_bilstm)
ax2.plot(fpr_bilstm, tpr_bilstm, label=f'Bi-LSTM (AUC = {auc_bilstm:.3f})')
ax2.plot([0, 1], [0, 1], 'k--')
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('ROC Curve - Bi-LSTM')
ax2.legend()

plt.tight_layout()
plt.show()
```

## 6. Gated Recurrent Unit (GRU)

Implement GRU for tabular data classification. GRU is computationally more efficient than LSTM while maintaining similar performance.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

print("="*60)
print("GATED RECURRENT UNIT (GRU)")
print("="*60)

class GRUNet(nn.Module):
    def __init__(self, input_dim):
        super(GRUNet, self).__init__()
        
        # GRU layers
        self.gru = nn.GRU(
            input_size=1,  # Each feature as a sequence element
            hidden_size=64,
            num_layers=2,
            batch_first=True,
            dropout=0.2,
            bidirectional=True
        )
        
        # After bidirectional GRU: hidden_size * 2
        gru_output_size = 64 * 2
        
        self.classifier = nn.Sequential(
            nn.Linear(gru_output_size, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        # Reshape: (batch_size, features) -> (batch_size, features, 1)
        x = x.unsqueeze(-1)
        
        # GRU forward pass
        gru_out, hidden = self.gru(x)
        
        # Take the last output
        gru_out = gru_out[:, -1, :]
        
        # Classification
        output = self.classifier(gru_out)
        return output

# Prepare data
device = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

X_train_gru = torch.FloatTensor(X_train).to(device)
y_train_gru = torch.FloatTensor(y_train).unsqueeze(1).to(device)
X_test_gru = torch.FloatTensor(X_test).to(device)
y_test_gru = torch.FloatTensor(y_test).unsqueeze(1).to(device)

# Create model
gru_model = GRUNet(X_train.shape[1]).to(device)

# Training setup
criterion = nn.BCELoss()
optimizer = optim.Adam(gru_model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=10, factor=0.5)

# Data loader
train_dataset_gru = TensorDataset(X_train_gru, y_train_gru)
train_loader_gru = DataLoader(train_dataset_gru, batch_size=32, shuffle=True)

# Training loop
num_epochs = 100
train_losses_gru = []
print("Training GRU...")

for epoch in range(num_epochs):
    gru_model.train()
    epoch_loss = 0.0
    
    for batch_X, batch_y in train_loader_gru:
        optimizer.zero_grad()
        outputs = gru_model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        
        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(gru_model.parameters(), max_norm=1.0)
        
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader_gru)
    train_losses_gru.append(avg_loss)
    scheduler.step(avg_loss)
    
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')

# Evaluation
gru_model.eval()
with torch.no_grad():
    test_outputs_gru = gru_model(X_test_gru)
    test_proba_gru = test_outputs_gru.cpu().numpy().ravel()
    test_pred_gru = (test_proba_gru >= 0.5).astype(int)

# Calculate metrics
y_test_np = y_test_gru.cpu().numpy().ravel()
accuracy_gru = accuracy_score(y_test_np, test_pred_gru)
auc_gru = roc_auc_score(y_test_np, test_proba_gru)

print(f"\nGRU Results:")
print(f"Accuracy: {accuracy_gru:.4f}")
print(f"ROC-AUC: {auc_gru:.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test_np, test_pred_gru))
print("\nClassification Report:")
print(classification_report(y_test_np, test_pred_gru))

# Plot results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Training loss
ax1.plot(train_losses_gru)
ax1.set_title('GRU Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')

# ROC curve
fpr_gru, tpr_gru, _ = roc_curve(y_test_np, test_proba_gru)
ax2.plot(fpr_gru, tpr_gru, label=f'GRU (AUC = {auc_gru:.3f})')
ax2.plot([0, 1], [0, 1], 'k--')
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('ROC Curve - GRU')
ax2.legend()

plt.tight_layout()
plt.show()
```

## 8. TabTransformer

Implement TabTransformer, which applies the Transformer architecture to tabular data by treating each feature as a token.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import math

print("="*60)
print("TABTRANSFORMER")
print("="*60)

class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        super(MultiHeadAttention, self).__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.d_k = d_model // num_heads
        
        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.W_o = nn.Linear(d_model, d_model)
        
    def forward(self, query, key, value):
        batch_size = query.size(0)
        
        # Linear transformations
        Q = self.W_q(query).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        K = self.W_k(key).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        V = self.W_v(value).view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)
        
        # Scaled dot-product attention
        scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(self.d_k)
        attention_weights = torch.softmax(scores, dim=-1)
        attention = torch.matmul(attention_weights, V)
        
        # Concatenate heads
        attention = attention.transpose(1, 2).contiguous().view(batch_size, -1, self.d_model)
        
        # Output linear transformation
        output = self.W_o(attention)
        return output

class TransformerBlock(nn.Module):
    def __init__(self, d_model, num_heads, d_ff, dropout=0.1):
        super(TransformerBlock, self).__init__()
        self.attention = MultiHeadAttention(d_model, num_heads)
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        
        self.feed_forward = nn.Sequential(
            nn.Linear(d_model, d_ff),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_ff, d_model)
        )
        
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        # Self-attention
        attn_output = self.attention(x, x, x)
        x = self.norm1(x + self.dropout(attn_output))
        
        # Feed forward
        ff_output = self.feed_forward(x)
        x = self.norm2(x + self.dropout(ff_output))
        
        return x

class TabTransformer(nn.Module):
    def __init__(self, input_dim, d_model=128, num_heads=8, num_layers=4, d_ff=256, dropout=0.1):
        super(TabTransformer, self).__init__()
        
        # Input embedding (treat each feature as a token)
        self.input_embedding = nn.Linear(1, d_model)
        self.positional_encoding = nn.Parameter(torch.randn(1, input_dim, d_model))
        
        # Transformer blocks
        self.transformer_blocks = nn.ModuleList([
            TransformerBlock(d_model, num_heads, d_ff, dropout)
            for _ in range(num_layers)
        ])
        
        # Classification head
        self.classifier = nn.Sequential(
            nn.Linear(d_model * input_dim, 256),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        batch_size = x.size(0)
        
        # Reshape and embed: (batch_size, features) -> (batch_size, features, 1) -> (batch_size, features, d_model)
        x = x.unsqueeze(-1)  # Add feature dimension
        x = self.input_embedding(x)  # Embed each feature
        x = x + self.positional_encoding  # Add positional encoding
        
        # Apply transformer blocks
        for transformer in self.transformer_blocks:
            x = transformer(x)
        
        # Flatten and classify
        x = x.view(batch_size, -1)
        output = self.classifier(x)
        
        return output

# Prepare data
device = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

X_train_tabtrans = torch.FloatTensor(X_train).to(device)
y_train_tabtrans = torch.FloatTensor(y_train).unsqueeze(1).to(device)
X_test_tabtrans = torch.FloatTensor(X_test).to(device)
y_test_tabtrans = torch.FloatTensor(y_test).unsqueeze(1).to(device)

# Create model
tabtrans_model = TabTransformer(
    input_dim=X_train.shape[1],
    d_model=64,  # Reduced for smaller datasets
    num_heads=4,
    num_layers=3,
    d_ff=128,
    dropout=0.1
).to(device)

# Training setup
criterion = nn.BCELoss()
optimizer = optim.Adam(tabtrans_model.parameters(), lr=0.0001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=10, factor=0.5)

# Data loader
train_dataset_tabtrans = TensorDataset(X_train_tabtrans, y_train_tabtrans)
train_loader_tabtrans = DataLoader(train_dataset_tabtrans, batch_size=16, shuffle=True)  # Smaller batch size

# Training loop
num_epochs = 100
train_losses_tabtrans = []
print("Training TabTransformer...")

for epoch in range(num_epochs):
    tabtrans_model.train()
    epoch_loss = 0.0
    
    for batch_X, batch_y in train_loader_tabtrans:
        optimizer.zero_grad()
        outputs = tabtrans_model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        
        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(tabtrans_model.parameters(), max_norm=1.0)
        
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader_tabtrans)
    train_losses_tabtrans.append(avg_loss)
    scheduler.step(avg_loss)
    
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')

# Evaluation
tabtrans_model.eval()
with torch.no_grad():
    test_outputs_tabtrans = tabtrans_model(X_test_tabtrans)
    test_proba_tabtrans = test_outputs_tabtrans.cpu().numpy().ravel()
    test_pred_tabtrans = (test_proba_tabtrans >= 0.5).astype(int)

# Calculate metrics
y_test_np = y_test_tabtrans.cpu().numpy().ravel()
accuracy_tabtrans = accuracy_score(y_test_np, test_pred_tabtrans)
auc_tabtrans = roc_auc_score(y_test_np, test_proba_tabtrans)

print(f"\nTabTransformer Results:")
print(f"Accuracy: {accuracy_tabtrans:.4f}")
print(f"ROC-AUC: {auc_tabtrans:.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test_np, test_pred_tabtrans))
print("\nClassification Report:")
print(classification_report(y_test_np, test_pred_tabtrans))

# Plot results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Training loss
ax1.plot(train_losses_tabtrans)
ax1.set_title('TabTransformer Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')

# ROC curve
fpr_tabtrans, tpr_tabtrans, _ = roc_curve(y_test_np, test_proba_tabtrans)
ax2.plot(fpr_tabtrans, tpr_tabtrans, label=f'TabTransformer (AUC = {auc_tabtrans:.3f})')
ax2.plot([0, 1], [0, 1], 'k--')
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('ROC Curve - TabTransformer')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Create a comprehensive comparison of all models
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, roc_auc_score

# Dictionary to store model results
model_results = {}

# Function to evaluate and store model performance
def evaluate_model(model, model_name, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
    
    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba) if y_proba is not None else None
    
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'ROC-AUC': auc
    }

# Collect results from all models (if they exist in the current session)
models_to_evaluate = [
    (classifier, 'Logistic Regression'),
    (rf, 'Random Forest'),
    (svm, 'SVM'),
]

# Add gradient boosting if it exists
try:
    if 'gb' in locals():
        models_to_evaluate.append((gb, 'Gradient Boosting'))
except:
    pass

# Add XGBoost if it exists
try:
    if 'xgb_model' in locals():
        models_to_evaluate.append((xgb_model, 'XGBoost'))
except:
    pass

# Add CatBoost if it exists
try:
    if 'catboost_model' in locals():
        models_to_evaluate.append((catboost_model, 'CatBoost'))
except:
    pass

# Add LightGBM if it exists
try:
    if 'lgb_model' in locals():
        models_to_evaluate.append((lgb_model, 'LightGBM'))
except:
    pass

# Add AdaBoost if it exists
try:
    if 'ada_model' in locals():
        models_to_evaluate.append((ada_model, 'AdaBoost'))
except:
    pass

# Add Extra Trees if it exists
try:
    if 'et_model' in locals():
        models_to_evaluate.append((et_model, 'Extra Trees'))
except:
    pass

# Add PyTorch model results manually (since they're not sklearn models)
advanced_models = [
    ('PyTorch Neural Network', 'test_accuracy', 'test_auc'),
    ('Stacking Ensemble', 'accuracy_stacking', 'auc_stacking'),
    ('CNN', 'accuracy_cnn', 'auc_cnn'),
    ('Deep MLP', 'accuracy_mlp', 'auc_mlp'),
    ('ANN', 'accuracy_ann', 'auc_ann'),
    ('Bi-LSTM', 'accuracy_bilstm', 'auc_bilstm'),
    ('GRU', 'accuracy_gru', 'auc_gru'),
    ('TabPFN', 'accuracy_tabpfn', 'auc_tabpfn'),
    ('TabTransformer', 'accuracy_tabtrans', 'auc_tabtrans')
]

for model_name, acc_var, auc_var in advanced_models:
    try:
        if acc_var in locals() and auc_var in locals():
            acc_val = locals()[acc_var]
            auc_val = locals()[auc_var]
            if acc_val is not None and auc_val is not None:
                result = {
                    'Model': model_name,
                    'Accuracy': acc_val,
                    'ROC-AUC': auc_val
                }
                model_results[model_name] = result
    except:
        pass

# Evaluate all sklearn models
for model, name in models_to_evaluate:
    try:
        result = evaluate_model(model, name, X_test, y_test)
        model_results[name] = result
        print(f"{name}: Accuracy = {result['Accuracy']:.4f}, ROC-AUC = {result['ROC-AUC']:.4f}")
    except Exception as e:
        print(f"Error evaluating {name}: {e}")

# Create comparison DataFrame
if model_results:
    comparison_df = pd.DataFrame(list(model_results.values()))
    comparison_df = comparison_df.sort_values('ROC-AUC', ascending=False)
    
    print("\n" + "="*50)
    print("MODEL PERFORMANCE COMPARISON")
    print("="*50)
    print(comparison_df.to_string(index=False, float_format='%.4f'))
    
    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Accuracy comparison
    ax1.barh(comparison_df['Model'], comparison_df['Accuracy'])
    ax1.set_xlabel('Accuracy')
    ax1.set_title('Model Accuracy Comparison')
    ax1.set_xlim(0, 1)
    
    # Add accuracy values on bars
    for i, v in enumerate(comparison_df['Accuracy']):
        ax1.text(v + 0.01, i, f'{v:.3f}', va='center')
    
    # ROC-AUC comparison
    ax2.barh(comparison_df['Model'], comparison_df['ROC-AUC'])
    ax2.set_xlabel('ROC-AUC')
    ax2.set_title('Model ROC-AUC Comparison')
    ax2.set_xlim(0, 1)
    
    # Add ROC-AUC values on bars
    for i, v in enumerate(comparison_df['ROC-AUC']):
        ax2.text(v + 0.01, i, f'{v:.3f}', va='center')
    
    plt.tight_layout()
    plt.show()
    
    # Find best model
    best_model = comparison_df.iloc[0]
    print(f"\n Best Model: {best_model['Model']}")
    print(f"   Accuracy: {best_model['Accuracy']:.4f}")
    print(f"   ROC-AUC: {best_model['ROC-AUC']:.4f}")
else:
    print("No models found for comparison. Please run the model training cells first.")
```
